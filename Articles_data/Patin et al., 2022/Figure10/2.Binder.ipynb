```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_led = glob('*BW3*LED*dE*')[0]
file_hpx = glob('*BW3*HPX*dE*')[0]

df_led = pd.read_csv(file_led, header = [0,1])
df_hpx = pd.read_csv(file_hpx, header = [0,1])

E_led = df_led['cum_MJ/m2']
E_hpx = df_hpx['cum_MJ/m2']

L_led = df_led['L*','mean']
L_hpx = df_hpx['L*','mean']
a_led = df_led['a*','mean']
a_hpx = df_hpx['a*','mean']
b_led = df_led['b*','mean']
b_hpx = df_hpx['b*','mean']

dL_led = L_led - L_led[0]
dL_hpx = L_hpx - L_hpx[0]
da_led = a_led - a_led[0]
da_hpx = a_hpx - a_hpx[0]
db_led = b_led - b_led[0]
db_hpx = b_hpx - b_hpx[0]

fig, ax = plt.subplots(1,1,figsize = (13,9))

#ax.plot(a_led,b_led, label = 'LED')
#ax.plot(a_hpx,b_hpx, label = 'HPX')
ax.plot(E_led, dL_led, label = 'LED')
ax.plot(E_hpx, dL_hpx, label = 'HPX')

plt.legend()
plt.tight_layout()
plt.show()
```
